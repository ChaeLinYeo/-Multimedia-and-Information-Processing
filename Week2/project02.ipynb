```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = np.zeros(shape=(512, 512), dtype=np.uint8)

print("img의 shape : ", img.shape)
print("img[0,0]의 값 : ", img[0,0])

plt.imshow(img, cmap='gray', vmin=0, vmax=255) #스트레칭이 알아서 되긴 하는데 정확한 값을 보려고 vmin, vmax를 지정해준다.
```

```python
for i in range(0,200):
    for j in range(0,200):
        img[i, j] = 120
#이중for문 자주 쓴다
print("img[0,0]의 값 : ", img[0,0])
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
img = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
print("img의 shape : ", img.shape)

plt.subplot(1, 3, 1)
plt.title("All zero")
plt.imshow(img)

for i in range(512):
    for j in range(512):
        img[i, j, 0] = 255
#RGB채널중에 R값만 255로 꽉 채워준다

plt.subplot(1, 3, 2)
plt.title("Red color")
plt.imshow(img)

for i in range(100):
    for j in range(100):
        img[i, j, 0] = 0
        img[i, j, 1] = 128
        img[i, j, 2] = 128
#G,B값만 절반씩 채워준다

#표만들기
plt.subplot(1, 3, 3)
plt.title("other color")
plt.imshow(img)
```

```python
flowers_RGB = cv2.imread("./Flowers.jpg", cv2. IMREAD_COLOR)#꽃 이미지 출력
flowers_RGB = cv2.cvtColor(flowers_RGB, cv2. COLOR_BGR2RGB)

print("영상의 data type : ", flowers_RGB.dtype)
print("영상의 shape : ", flowers_RGB.shape)
print("flowers_RGB[0,0]의 값 : ", flowers_RGB[0,0])

plt.imshow(flowers_RGB)
```

```python
plt.imshow(flowers_RGB)

flowers_R = flowers_RGB[:, :, 0]
flowers_G = flowers_RGB[:, :, 1]
flowers_B = flowers_RGB[:, :, 2]

print("flowers_RGB의 shape : ", flowers_RGB.shape)
print("flowers_R(단일 채널)의 shape : ", flowers_R.shape)

plt.figure()
plt.subplot(1, 3, 1)
plt.title("R channel")
plt.imshow(flowers_R, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("G channel")
plt.imshow(flowers_G, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("B channel")
plt.imshow(flowers_B, cmap='gray')
```

```python
lena_gray = cv2.imread("./Lena_gray.jpg", cv2. IMREAD_GRAYSCALE)

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(lena_gray, cmap = 'gray')

h, w = lena_gray.shape
lena_add = np.zeros((h,w), np.uint8)

#영상처리는 대부분 2중 for문으로 한다.
for i in range(h):
    for j in range(w):
        lena_add[i, j] = lena_gray[i, j] + 50

plt.subplot(1, 3, 2)
plt.title("Add 50")
plt.imshow(lena_add, cmap='gray')

lena_add = lena_gray + 100

plt.subplot(1, 3, 3)
plt.title("Add 100")
plt.imshow(lena_add, cmap='gray')
```

```python
lena_div = lena_gray / 2

plt.figure()
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(lena_gray, cmap='gray')

plt.subplot(1,2,2)
plt.title("Div")
plt.imshow(lena_div, cmap='gray', vmin=0, vmax=255)
```

```python
lena_gray = cv2.imread("./Lena_gray.jpg", cv2. IMREAD_GRAYSCALE)

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(lena_gray, cmap='gray')

lena_mul = np. int32(lena_gray)*2
lena_mul[lena_mul > 255] = 255
lena_mul = np.uint8(lena_mul)

plt.subplot(1, 3, 2)
plt.title("Mul (Saturated)")
plt.imshow(lena_mul, cmap='gray')

lena_sub = np. int32(lena_gray) - 100
lena_sub[lena_sub < 0] = 0
lena_sub = np. uint8(lena_sub)

plt.subplot(1, 3, 3)
plt.title("Sub (Saturated)")
plt.imshow(lena_sub, cmap='gray')
```
